```python
import pickle
import numpy as np
import fun_dual as f
from itertools import permutations
from sympy import Matrix
from IPython.display import display
```

# Print solution $Y^{(a,k)}$, $C_i$ and $Q_i$ of the SDP from Eq. (150)


```python
n,K,d=10,5,4

with open("inf_cert_10", 'rb') as inf_cert:
    Y,C,Q=pickle.load(inf_cert)

y=f.yijtp_fun(Y,n)

f4,_=f.f4_fun(n)

f2M,_=f.f2M_fun(n)

D=f.D_fun(C,n,d)

for a in range(n+1):
    for k in range(a,int((n+a)/2)+1):
        print("Y^(%s,%s)=" % (a,k))
        display(Matrix((Y[f2M[a,k]])))

print("C=")
display(Matrix(C))
print("Q=")
display(Matrix(Q))
```

# Objective value

```python
obj_val = (K*D[0]-C[0])+(2**n/K-1)*Q[0] - y[f4[0, 0, 0, 0]]/f.gamma(0, 0, 0, 0, n)

print("obj_val=%s" % obj_val)
```

# Maximum violation constraints

```python
const=[]

for i, j, t, p in f4.keys():

    if (t - p) % 2 == 0 and not i==0 and not j == 0 and not i == j == t == p:

        zero = sum([y[f4[a, b, c, q]] + Q[a+b-c-q]
                    for a, b, c, q in f4.keys() if t - p == c - q and (i, j, i + j - t - p) in 
                    list(permutations((a, b, a + b - c - q)))])

        const.append(abs(zero))


print("Structure constraints = %s" % max(const))

const=[]

for i in range(1, d):
    zero = K*D[i]-C[i] +(2**n/K-2)*Q[i]-Q[0] - y[f4[i, i, i, i]] - 2*y[f4[i, 0, 0, 0]]

    const.append(abs(zero))

print("Other constraints 1 = %s" % max(const))

const=[]

for i in range(d, n + 1):
    zero = K*D[i] +(2**n/K-2)*Q[i]-Q[0] - y[f4[i, i, i, i]] - 2*y[f4[i, 0, 0, 0]]
    const.append(abs(zero))
        
print("Other constraints 2 = %s" % max(const))
```

# Minimum eigenvalue

```python
ref = np.infty
for i in range(len(Y)):
    min_eig = min(np.linalg.eigvals(Y[i].astype('float')))
    if ref >= min_eig:
        ref = min_eig

print("Minimum eigenvalue = %s" % ref)
```
